```python
stones={
    1:"Garnet",
    2:"Amethyst",
    3:"Aquamarine",
    4:"Diamond",
    5:"Emerald",
    6:"Pearl",
    7:"Ruby",
    8:"Peridot",
    9:"Sapphire",
    10:"Opal",
    11:"Topaz",
    12:"Turquoise"
}


def user_interface():
    while True:
        Month = int(input("What is your birth month?"))
        if Month in stones.keys():
            return stones[Month]
        else:
            print("Invalid number. Please try again.")

print(user_interface())
```

```python
stones={
    1:"Garnet",
    2:"Amethyst",
    3:"Aquamarine",
    4:"Diamond",
    5:"Emerald",
    6:"Pearl",
    7:"Ruby",
    8:"Peridot",
    9:"Sapphire",
    10:"Opal",
    11:"Topaz",
    12:"Turquoise"
}

def birthstone():
    months_to_int = {
        'january': 1,
        'february': 2,
        'march': 3,
        'april': 4,
        'may': 5,
        'june': 6,
        'july': 7,
        'august': 8,
        'september': 9,
        'october': 10,
        'november': 11,
        'december': 12
    }
    month = input('What month were you born in? ').lower()
    while month not in months_to_int:
        print('That is not a valid month. Please try again.')
        month = input('What month were you born in? ').lower()
    return stones[months_to_int[month]]


birthstone()
```
